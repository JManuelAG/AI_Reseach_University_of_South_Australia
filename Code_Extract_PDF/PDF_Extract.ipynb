```python
import xml.etree.ElementTree as ET

def find_text_recursively(element):
    """Recursively find text within an element."""
    text = ' '
    if element.text:
        text += element.text.strip()
    for child in element:
        text += find_text_recursively(child)
    return text

def find_bbox_of_abstract(xml_file):
    abs_state = 0
    abstract = ""

    # Parse the XML file
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Iterate through all 'LTRect' elements
    for ltr in root.findall('.//LTTextBoxHorizontal'):
        # Aggregate text from all descendants
        aggregated_text = find_text_recursively(ltr)

        # Check if aggregated text contains 'Abstract'
        if 'Abstract' in aggregated_text:
            # Return the 'bbox' attribute of the LTRect element
            return aggregated_text

    return 'Abstract section not found'

# Path to your XML file
xml_file_path = './test.xml'
bbox_value = find_bbox_of_abstract(xml_file_path)
print(bbox_value)
```

```python
# access the data using coordinates
tittle = pdf.pq(f'LTTextBoxHorizontal:in_bbox("{bbox_value[1:-1]}")').text()

print(tittle)
```
